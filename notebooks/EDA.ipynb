```python
# EDA.ipynb
import sys
import os
import pandas as pd
import matplotlib.pyplot as plt

# Add the path to the scripts folder
sys.path.append(os.path.abspath(os.path.join('..', 'scripts')))


# Load the data
file_path = "../data/raw_analyst_ratings.csv"  # Update with the correct path if needed
data = pd.read_csv(file_path)
# eda_functions.py


styled_table = (
data.head(10) # Display the first 10 rows
.style
.set_table_styles(
[
{"selector": "thead", "props": [("background-color", "#4CAF50"), ("color", "white"), ("font-weight", "bold")]},
{"selector": "tr:hover", "props": [("background-color", "#f1f1f1")]},
]
)
.highlight_max(axis=0, color="lightgreen") # Highlight maximum values in columns
.highlight_min(axis=0, color="lightcoral") # Highlight minimum values in columns
.set_caption("Styled Table: Preview of raw_analyst_ratings Dataset") # Add a caption
)

# Display the styled table
styled_table

# Select numeric columns
numeric_columns = data.select_dtypes(include=["number"])

# Initialize a dictionary to hold the results
statistics_result = {}

# Calculate statistics for each numeric column
for column in numeric_columns.columns:
    statistics_result[column] = {
    "Mean": numeric_columns[column].mean(),
    "Median": numeric_columns[column].median(),
    "Mode": numeric_columns[column].mode().iloc[0] if not numeric_columns[column].mode().empty else None,
    "Min": numeric_columns[column].min(),
    "Max": numeric_columns[column].max(),
    "Standard Deviation": numeric_columns[column].std(),
    "Count": numeric_columns[column].count(),
    }

# Convert the result to a DataFrame for better display
stats_df = pd.DataFrame(statistics_result).transpose()

# Style the DataFrame
styled_stats_df = (
stats_df.style
.set_table_styles(
[
{"selector": "thead", "props": [("background-color", "#4CAF50"), ("color", "white"), ("font-weight", "bold")]},
{"selector": "tr:hover", "props": [("background-color", "#f1f1f1")]},
]
)
.set_caption("Statistics for Numeric Columns")
.background_gradient(cmap="viridis", subset=["Mean", "Median", "Min", "Max", "Standard Deviation"])
.format("{:.2f}", subset=["Mean", "Median", "Min", "Max", "Standard Deviation"])
)

# Display the styled table
styled_stats_df

```
